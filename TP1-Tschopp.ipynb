```python
import pandas as pd

import duckdb
```

```python
# Especifica la ruta del archivo CSV en una variable
DIR = r'G:\Mi unidad\01-Maestria Ciencia de Datos\DMEyF'

# Construye la ruta completa al archivo
file_path = f'{DIR}\competencia_01_crudo.csv'

# Carga el archivo CSV en un DataFrame de pandas
df = pd.read_csv(file_path)
```

```python
# Muestra los primeros cinco registros del DataFrame
df.head()
```

```python
# Función para determinar el estado del cliente en el tiempo
def valor_para_foto_mes(foto_mes, numero_de_cliente, df):
    foto_mes_plus_1 = foto_mes + 1
    foto_mes_plus_2 = foto_mes + 2
    
    # Verifica si existe un registro para (foto_mes + 1)
    if not df[(df['foto_mes'] == foto_mes_plus_1) & (df['numero_de_cliente'] == numero_de_cliente)].empty:
        # Verifica si existe un registro para (foto_mes + 2)
        if not df[(df['foto_mes'] == foto_mes_plus_2) & (df['numero_de_cliente'] == numero_de_cliente)].empty:
            return 'CONTINUA'
        else:
            return 'BAJA+1'
    else:
        # Si no existe para (foto_mes + 1), entonces podría ser BAJA+2 si falta también (foto_mes + 2)
        if not df[(df['foto_mes'] == foto_mes_plus_2) & (df['numero_de_cliente'] == numero_de_cliente)].empty:
            return 'BAJA+1'
        else:
            return 'BAJA+2'

# Crear nuevas columnas para cada valor único de 'foto_mes'
for mes in df['foto_mes'].unique():
    nueva_columna = f'foto_mes_{mes}'
    
    # Aplicar la función a cada registro para llenar la nueva columna
    df[nueva_columna] = df.apply(lambda x: valor_para_foto_mes(x['foto_mes'], x['numero_de_cliente'], df) if x['foto_mes'] == mes else None, axis=1)

# Mostrar el DataFrame resultante
df
```

```python
# Crear o conectar a una base de datos en un archivo llamado 'DMEyF2024.duckdb'
# Para una base de datos en memoria, usa ':memory:' en lugar del nombre del archivo
conn = duckdb.connect('DMEyF2024.duckdb')

# Especifica la ruta del archivo CSV en una variable
DIR = r'G:\Mi unidad\01-Maestria Ciencia de Datos\DMEyF'

# Construye la ruta completa al archivo
csv_file_path = f'{DIR}\competencia_01_crudo.csv'

# Cargar el archivo CSV en una tabla llamada 'competencia'
conn.execute(f"""
    CREATE TABLE competencia AS
    SELECT * FROM read_csv_auto('{csv_file_path}')
""")

# Verificar los primeros registros para asegurarse de que la carga fue exitosa
df = conn.execute("SELECT * FROM competencia LIMIT 5").fetchdf()
print(df)

# Cerrar la conexión
conn.close()
```
